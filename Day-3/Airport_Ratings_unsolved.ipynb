```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import os
from pprint import pprint

# Import API key
f = open(os.path.join("..","..","..","gmaps_api_key.txt"), "r")
api_key = f.read()
f.close()
```

```python
# Import cities file into the cities_pd DataFrame
cities_pd = pd.read_csv("Cities.csv")
cities_pd.head()
```

```python
# Add columns for lat, lng, airport name, airport address, airport rating.
# Hint: Be sure to specify "" as the initial value for each column.
cities_pd["lat"] = ""
cities_pd["lng"] = ""
cities_pd.head()
```

```python
# Loop through the cities_pd and get the lat/long for each city
# Hint: `requests.get(target_url).json()`
# Hint: `for index, row in cities_pd.iterrows():`
# Hint: `cities_pd.loc`

base_url = "https://maps.googleapis.com/maps/api/geocode/json"
params = {
    "address": "",
    "key": api_key,
}

for index, row in cities_pd.iterrows():
    params["address"] = row["City"] + ", " + row['State']
    print(f"\nGetting the info for {row['City']},{row['State']}\n{'-'*30}")
    response = requests.get(base_url, params=params)
    responseJson = response.json()
    cities_pd.loc[index, 'lat'] = responseJson["results"][0]["geometry"]["location"]["lat"]
    cities_pd.loc[index, 'lng'] = responseJson["results"][0]["geometry"]["location"]["lng"]
    print(f"{row['City']}, {row['State']}'s coordinates are {cities_pd.loc[index, 'lat']}, {cities_pd.loc[index, 'lng']}")
    
```

```python
cities_pd.head()
```

```python
# Use the lat/lng we recovered to identify airports
# Hint: Code is similar to above block (i.e. you will need to use cities_pd.iterrows() again)
# Hint: Use the keyword search for "international+airport"
# Hint: Use a try-except to handle airports that are missing ratings
airports_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "location": "",
    "radius": 8000,
    "type": "airport",
    "key": api_key,
    "keyword": 'international airport'
}

cities_pd['nearest airport'] = ""

for index, row in cities_pd.iterrows():
    params['location'] = f"{row['lat']},{row['lng']}"
    print(f"\nGetting the {row['City']},{row['State']}'s nearest airport'\n{'-'*30}")
    response = requests.get(airports_url, params=params)
    responseJson = response.json()
    
    try:
        #print(responseJson)
        print(f"Closest airport to {row['City']}, {row['State']} is {responseJson['results'][0]['name']}.")
        cities_pd.loc[index, 'nearest airport'] = responseJson['results'][0]['name'] 
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```
