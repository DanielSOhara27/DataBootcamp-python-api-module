```python
# Dependencies
import requests
import json
import os
from pprint import pprint

p = os.path.join("..","..","..","openWeather_api_key.txt")
f = open(p, "r")
api_key = f.read()
f.close()
```

```python
# Build query URL and request your results in Celsius
url = "http://api.openweathermap.org/data/2.5/forecast?"
url = url +"APPID="+api_key+"&q=Bujumbura&units=metric"
# Get weather data
response = requests.get(url)
response
```

```python
# Get temperature from JSON response
jsonWeather = response.json()
pprint(jsonWeather)

temp = ((jsonWeather.get("list")[0]).get("main")).get("temp")
temp
```
